**DATA FRAME - OBLICZANIE NOWYCH KOLUMN**

```python
df = fetch_financial_data('UBER')
df.info()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
df.describe()
```

```python
#dodawanie nowych kolumn
df['avg'] = (df.Open + df.Close)/2
df
```

```python
#sortowanie
df = df.sort_index()
df
```

```python
df['Shift'] = df[['Close']].shift(1)
df['diff_day'] = ((df.Close / df.Shift) - 1) * 100

df
```

```python
import seaborn as sns
sns.set()
df[['diff_day']].hist(bins = 100)
```

**FILTROWANIE DANYCH**

```python
df_positive = df [df.diff_day > 0]
```

```python
df_negative = df [df.diff_day < 0]
```

```python
#średnia dzienna dodatnia stopa zwrotu
df_positive.diff_day.mean()
```

```python
#średnia ujemna stopa zwrotu
df_negative.diff_day.mean()
```

```python
df [df.Close == df.High]
```

```python
df [df.Close == df.Low]
```

```python
#łączenie warunków logicznych
df[((df.index > '2020-01-01') & (df.index < '2020-02-02') & (df.diff_day > 0)) | (df.diff_day < 0)]
```

```python
#wyciaganie miesiacow, lat, dni
df[(df.index.year == 2022) & ((df.index.month == 2) | (df.index.month == 1)) & ((df.index.day == 1) | (df.index.day == 3))]
```
