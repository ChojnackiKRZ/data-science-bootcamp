**DATA FRAME - OBLICZANIE NOWYCH KOLUMN**

```python
df = fetch_financial_data('UBER')
df.info()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
df.describe()
```

```python
#dodawanie nowych kolumn
df['avg'] = (df.Open + df.Close)/2
df
```

```python
#sortowanie
df = df.sort_index()
df
```

```python
df['%change'] = (1 - (df[['Close']].shift(1) / df[['Close']])) * 100
df
```

```python
import seaborn as sns
sns.set()
df[['%change']].hist(bins = 100)
```
